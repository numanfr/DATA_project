```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import os
```

```python
PATH = 'individual_book_train'
files = [i for i in os.listdir(PATH) if i[-4:] == ".csv"]
```

```python
data = []
for i,file in enumerate(files[:]):
    print(i/len(files))
    df = pd.read_csv(f"{PATH}/{file}")
    tids = df['time_id'].unique()
    for tid in tids[:]:
        prices = df[df["time_id"] == tid]["bid_price1"].to_numpy()
        prices = np.abs(prices[:-1] - prices[1:])
        price = prices[prices>0]
        if len(price) > 0:
            
            price = 0.01/price.min()
            data.append([file,tid,price])
        else:
            data.append([file,tid,np.NAN])
        
```
